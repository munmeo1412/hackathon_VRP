```
dt = pd.read_csv('answer.csv', header=None,names=['driver_id','order_code','lat','long','weight','distance','deadline_pickup','priority'], skiprows = 1)
dt.info()
```

```
# dt.reset_index(inplace=True, drop=True)
dt.head()
```

```

location = 10.781703, 106.661044
m = folium.Map(location=location,zoom_start=15)

for i in range(0,len(dt)):
       
    popup = folium.Popup('order_code', parse_html=True) 
    folium.Marker([dt['lat'].iloc[i],dt['long'].iloc[i]],popup=popup).add_to(m)
m
```

```
location = dt['lat'].mean(), dt['long'].mean()
m = folium.Map(location=location,zoom_start=15)

#The num of casulaties for each accident can be determined and the colour assigned then added to the basemap.
for i in range(0,len(dt)):
    num_of_driver = dt['driver_id'].iloc[i]
    if num_of_driver == 0:
        color = 'blue'
    elif num_of_driver == 1:
        color = 'green'
    elif num_of_driver == 2:
        color = 'pink'
    elif num_of_driver == 3:
        color = 'black'       
    else:
        color = 'red'
    
    popup = folium.Popup(dt['order_code'].iloc[i] + "_" + dt['priority'].astype(str).iloc[i], parse_html=True) 
    folium.Marker([dt['lat'].iloc[i],dt['long'].iloc[i]],popup=popup,icon=folium.Icon(color=color, icon='remove-circle')).add_to(m)
```
